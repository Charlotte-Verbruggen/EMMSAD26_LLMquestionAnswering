```python
# RQ2.2: compare size > friedman test per threshold and format in set A (significant difference between PlantUML and PlantUMLx10 and between JSON and JSONx10 for all thresholds)
```

```python
#only for plantUML:
df_A_25_PlantUML = df_A_25.loc[:,["PlantUML","PlantUMLx2","PlantUMLx10"]]
df_A_50_PlantUML = df_A_50.loc[:,["PlantUML","PlantUMLx2","PlantUMLx10"]]
df_A_75_PlantUML = df_A_75.loc[:,["PlantUML","PlantUMLx2","PlantUMLx10"]]
df_A_100_PlantUML = df_A_100.loc[:,["PlantUML","PlantUMLx2","PlantUMLx10"]]


df_A_25_PlantUML
```

```python
stat, p = friedmanchisquare(*[df_A_25_PlantUML[col] for col in df_A_25_PlantUML.columns])

# get dimensions
n = df_A_25_PlantUML.shape[0]
k = df_A_25_PlantUML.shape[1]

# effect size
W = stat / (n * (k - 1))

print("SET A - TRHESHOLD: 25% for all size of plantUML")
print("Friedman statistic:", stat)
print("p-value:", f"{p:.6f}")
print("Kendall's W:", W)
```
